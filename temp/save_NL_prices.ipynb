```python
import os
import pandas as pd

file_name = 'combined_data_cleaned_v3.csv'

# Getting the file path
notebook_dir = os.getcwd()
python_dir = os.path.dirname(notebook_dir)
data_folder = os.path.join(python_dir, 'Data')
data_file = os.path.join(data_folder, file_name)

# Reading the data as a DataFrame
df = pd.read_csv(data_file, decimal = ",")
```

```python
file_name = 'NL_prices.csv'

data_file = os.path.join(data_folder, file_name)

# Reading the data as a DataFrame
df_NL = pd.read_csv(data_file, decimal = ",", sep = ";", header = None)

df_NL.columns = ['NL']

NL_list = df_NL['NL'].values.tolist()
```
